```python
import pandas as pd
import os
```

```python
ruta = r"C:\Users\Ian\Documents\GitHub\Data_Pensiones\Descargas"
contenido = os.listdir(ruta)
archivos = []
for fichero in contenido:
    if os.path.isfile(os.path.join(ruta, fichero)) and fichero.endswith('.xls'):
        archivos.append(fichero)
archivos
```

```python
afiliados_fts=[]
for i in archivos:
    indice = i[:2]
    if indice == "1.":
        #print(i)
        afiliados_fts.append(i)
        print()
afiliados_fts
```

```python
#Tipo 1
diccionario = {}
salida = []
consol = []
n = 0
for t in afiliados_fts:
    df = pd.read_html(r"C:/Users/Ian/Documents/GitHub/Data_Pensiones/Descargas/"+t)
    df = pd.DataFrame(df[1], header=1)
    df = df.rename(columns={'TIPO Y SEXO':'Tipo y Sexo',
                   'FONDO TIPO AMÁS RIESGOSO':'Tipo A',
                   'FONDO TIPO BRIESGOSO':"Tipo B",
                   'FONDO TIPO CINTERMEDIO':'Tipo C',
                   'FONDO TIPO DCONSERVADOR':'Tipo D',
                   'FONDO TIPO EMÁS CONSERVADOR':'Tipo E'})
    for i in df["Tipo y Sexo"]:
        if("-" not in i):
            diccionario["afiliacion"]= i
        if("- DEPENDIENTES" in i):
            i = i[2:]
            diccionario["dependencia"]=i
        if("- INDEPENDIENTES" in i):
            i = i[2:]
            diccionario["dependencia"]=i
        if("--" in i):
            i = i[3:]
            diccionario["sexo"]=i
            diccionario["tipo A"]=df["Tipo A"][n]
            diccionario["tipo B"]=df["Tipo B"][n]
            diccionario["tipo C"]=df["Tipo C"][n]
            diccionario["tipo D"]=df["Tipo D"][n]
            diccionario["tipo E"]=df["Tipo E"][n]
            #print(diccionario)
            salida.append(diccionario.copy())
        n = n+1
    dfsalida = pd.DataFrame(salida)
    dfsalida = dfsalida[dfsalida["afiliacion"]!="TOTAL"]
    consol.append(dfsalida.copy())
pd.DataFrame(consol)
```
